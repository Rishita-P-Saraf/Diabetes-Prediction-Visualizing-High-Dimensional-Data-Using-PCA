```python
df=df.drop(columns= ['Outcome'],axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
dfscaled= scaler.fit_transform(df)
```

```python
dfscaled= pd.DataFrame(dfscaled)
dfscaled
```

```python
dfscaled.describe()
```

```python
dfscaled.shape
```

## 2. Covariance Matrix

```python
covMat= dfscaled.T @ dfscaled / 767
covMat
```

There are 8 features so 8 eigenvalues n 8 eigenvectors

## 3. Evaluate eigenValues and eigenVectors using covMat

```python
eigenValues, eigenVectors= np.linalg.eig(covMat)
```

```python
eigenValues
```

```python
eigenVectors
```

## 4. Finding PCs

```python
PC1 = dfscaled @ eigenVectors[:,0]
PC1
```

```python
PC2 = dfscaled @ eigenVectors[:,1]
PC2
```

```python
PC3 = dfscaled @ eigenVectors[:,7]
PC3
```

## PCA implementation using sklearn Library

```python
from sklearn.decomposition import PCA
pca= PCA(n_components= 8)
pca.fit_transform(dfscaled)
```

```python
from sklearn.decomposition import PCA
pca= PCA(n_components= 8)
PC = pca.fit_transform(dfscaled)
```

```python
pca.explained_variance_ratio_ 
```

```python
import matplotlib.pyplot as plt
plt.plot(np.cumsum(pca.explained_variance_ratio_ ))
plt.xlabel('Dimensions')
plt.ylabel('Explained Variance Ratio')
plt.savefig("ScreenPlot.png")
plt.show()
```
